```python
#Checking the missing values
df.isnull().sum()
```

```python
#reviewing na dthinking by taking a minute for replacing the numm values  by either mean or median median because it is less afttected by outliers so filling with median of entire column
# finding the median reviiew rating for each catogry and replacing because the review rating for every prooduct including the clothing and footwwear will aafftect overall insight and will create bias in the data 
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Snakecasing is done for converting the naing of columns into small letter with underscore dor better code consitency and vulnerablle to errors
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a new column age
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
#qcut splits the ages into 4 groups based on data distribution
```

```python
df[['age','age_group']].head(10)
```

```python
# purchase frequency_days another feature(column)
# we created dictionary frequency_mapping that maps eaach texual frequency to the eqivalent  number of days
# we will use map function to replace the text in column in the corresponding number of dayss
frequency_mapping={
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Anually' : 365,
    'Average 3 months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
